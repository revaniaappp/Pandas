<h1 style="font-size: 2.0em;">Seleksi baris dengan between()</h1>

<h1 style="font-size: 1.6em;">Import Modules</h1>

```python
import pandas as pd
import numpy as np
print(pd.__version__)
print(np.__version__)
```

<h1 style="font-size: 1.6em;">Persiapan Data Frame</h1>

```python
n_rows = 10
n_cols = 5
cols = tuple('ABCDE')
df = pd.DataFrame(np.random.randint(1,10,size=(n_rows,n_cols)), 
                  columns=cols)
df
```

<h1 style="font-size: 1.6em;">Seleksi baris dengan between()</h1>

```python
df[df['A'].between(2,5)]
```

```python
df[df['A'].between(2, 5, inclusive='left')]
```

```python
df[df['A'].between(2, 5, inclusive='right')]
```

```python
df[df['A'].between(2, 5, inclusive='both')]
```

```python
df[df['A'].between(2, 5, inclusive='neither')]
```
