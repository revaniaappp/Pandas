<h1 style="font-size: 2.0em;">Nilai Kumulatif pada Pandas Data Frame</h1>

<h1 style="font-size: 1.6em;">Import Modules</h1>

```python
import pandas as pd
print(pd.__version__)
```

<h1 style="font-size: 1.6em;">Persiapan Data Frame</h1>

```python
d = {'pemain':['Budi','Joni','Iwan','Budi','Budi','Iwan','Asep','Joni'], 
     'goal':[2,1,3,1,1,2,2,3]}
df = pd.DataFrame(d)
df
```

<h1 style="font-size: 1.6em;">Menampilkan nilai kumulatif (cummulative)</h1>

```python
df['goal'].cumsum().to_frame()
```

```python
df['jumlah_goal_kumulatif'] = df['goal'].cumsum()
df
```

```python
df['jumlah_goal_kumulatif_tiap_pemain'] = df.groupby('pemain')['goal'].cumsum()
df
```

```python
df['cummax'] = df['goal'].cummax()
df
```

```python
df['cummin'] = df['goal'].cummin()
df
```

```python
df['cumprod'] = df['goal'].cumprod()
df
```
