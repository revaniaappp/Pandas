<h1 style="font-size: 2.0em;">Plotting dari suatu Pandas Data Frame</h1>

<h1 style="font-size: 1.6em;">Import modules</h1>

```python
import pandas as pd
import numpy as np
print(pd.__version__)
print(np.__version__)
```

<h1 style="font-size: 1.6em;">Persiapan Data Frame</h1>

```python
n_rows = 40
n_cols = 5
cols = tuple('ABCDE')
df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), 
                  columns=cols)

df.head()
```

<h1 style="font-size: 1.6em;">Line Plot</h1>

```python
df.head()
```

```python
df.plot(kind='line')
```

```python
df[['A', 'B']].plot(kind='line')
```

<h1 style="font-size: 1.6em;">Bar Plot</h1>

```python
df.head()
```

```python
df.plot(kind='bar')
```

```python
df[['A', 'B']].plot(kind='bar')
```

```python
df[['A', 'B']].head().plot(kind='bar')
```

```python
df[['A', 'B']].head().plot(kind='barh')
```
