<h1 style="font-size: 2.0em;">Mengenal Cummulative Count dengan cumcount()</h1>

<h1 style="font-size: 1.6em;">Import Modules</h1>

```python
import pandas as pd

print(pd.__version__)
```

<h1 style="font-size: 1.6em;">Persiapan Data Frame</h1>

```python
d = {'penjual':['bejo', 'tejo', 'wati', 'bejo', 'cecep', 'tejo', 'wati', 'bejo'], 
     'barang':['monitor', 'monitor', 'keyboard', 'mouse', 'keyboard', 'monitor', 'laptop', 'monitor']}

df = pd.DataFrame(d)
df
```

<h1 style="font-size: 1.6em;">Mengenal Cummulative Count dengan cumcount()</h1>

```python
df['count_tiap_penjual'] = df.groupby('penjual').cumcount() + 1
df
```

```python
df['count_tiap_barang'] = df.groupby('barang').cumcount() + 1
df
```

```python
df['count_pasangan_kolom'] = df.groupby(['penjual', 'barang']).cumcount() + 1
df
```
